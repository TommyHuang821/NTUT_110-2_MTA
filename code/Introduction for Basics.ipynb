### 轉置矩陣(Transpose Matrix)

```python
Xt=np.transpose(X)
print('X transpose:\n{}'.format(Xt))
Xtt = np.transpose(Xt)
print('Xt transpose:\n{}'.format(Xtt))
```

```python
X = np.array([[150, 160],[50, 60]])
print('X:\n{}'.format(X))
detX=np.linalg.det(X)
detXt=np.linalg.det(np.transpose(X))
print('det(X):\n{}'.format(detX))
print('det(Xt):\n{}'.format(detXt))
```

```python
X = np.array([[1,2,3,4],[5,6,7,8]])
Y = np.array([[2,4,6,8],[1,3,5,7]])
XpYt = np.transpose(X+Y)
print('(X+Y)t:\n{}'.format(XpYt))
print('Xt+Yt:\n{}'.format(np.transpose(X)+np.transpose(Y)))
```

```python
X = np.array([[1,2,3],[5,6,7]])
Y = np.array([[2,4],[1,3],[5,7]])
print('X, size={}:\n{}'.format(X.shape,X))
print('Y, size={}:\n{}'.format(Y.shape,Y))
XY = np.dot(X,Y)
Xt = np.transpose(X)
Yt = np.transpose(Y)
print('(XY)t:\n{}'.format(np.transpose(XY)))
print('YtXt:\n{}'.format(np.dot(Yt,Xt)))
```

### 矩陣的秩

```python
X = np.array([[1,2,3],[2,4,6],[5,7,1]])
print('X:\n{}'.format(X))
```

```python
rankX=np.linalg.matrix_rank(X)
print('rank(X) = {}'.format(rankX))
```

### 逆矩陣

```python
A = np.array([[3,1],[4,2]])
print('A:\n{}'.format(A))
invA = np.linalg.inv(A)
print('inverse(A):\n{}'.format(invA))
```

### 特徵分解

```python
A = np.array([[2,1],[5, 6]])
print('A:\n{}'.format(A))
eigenvalue, eigenvector = np.linalg.eig(A)
print('eigenvalue of A:\n{}'.format(eigenvalue))
print('eigenvector of A:\n{}'.format(eigenvector))
```

```python
print(np.linalg.norm(eigenvector[:,0]))
print(np.linalg.norm(eigenvector[:,1]))

```

```python
v=[1,5]
L=np.linalg.norm(v)
v/L
```

### SVD

```python
A = np.array([[3,2,2],[2,3,-2]])
s,v,d = np.linalg.svd(A)
print('s:\n{}'.format(s))
print('v:\n{}'.format(v))
print('d:\n{}'.format(d))
```

# 第 2 章 機器學習相關機率論

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
np.random.seed(202202)

# using Scipy to plot Normal (pdf)
def draw_normal(mu, sigma, size):
    dnormal = np.random.normal(mu, sigma, size)
    _, bins_edge, _ = plt.hist(dnormal, bins=50, density=True, alpha=0.5)
    # pdf
    y = scipy.stats.norm.pdf(bins_edge, mu, sigma)
    plt.plot(bins_edge, y, label='$N(\overline {x} $=%.1f, $s$=%.1f)'%(mu, sigma))
    plt.xlabel('$x$(cm)')
    plt.ylabel('Probability')
    plt.title('Histogram of Normal distribution')

draw_normal(mu=175, sigma=5, size=1000)
plt.legend(loc=0, ncol=1)
plt.show()
```

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
np.random.seed(202202)

# using Scipy to plot Normal (pdf)
def draw_normal(mu, sigma, size, color='k'):
    dnormal = np.random.normal(mu, sigma, size)
    _, bins_edge, _ = plt.hist(dnormal, bins=50, density=True, alpha=0)
    # pdf
    y = scipy.stats.norm.pdf(bins_edge, mu, sigma)
    plt.plot(bins_edge, y, label='$N(\mu$=%.1f, $\sigma$=%.1f)'%(mu, sigma),color=color)
draw_normal(mu=0, sigma=1, size=1000, color='red')
draw_normal(mu=6, sigma=2, size=1000, color='blue')
# plt.legend(loc=0, ncol=1)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend(loc=0, ncol=1)
plt.show()
```

# 第 3 章 機器學習常用的統計學

## 相關係數範例

```python
import numpy as np
# example 1
X = np.array([180,175,170,165,160])
Y = np.array([80,70,70,70,50])
mu_x = np.mean(X)
mu_y = np.mean(Y)
X_zerooffset = X-mu_x
Y_zerooffset = Y-mu_y
Cov_XY = np.mean(X_zerooffset*Y_zerooffset)
sigma_X = np.mean(X_zerooffset**2)
sigma_Y = np.mean(Y_zerooffset**2)
r_XY = Cov_XY/np.sqrt(sigma_X*sigma_Y)
print('Cov_XY:\n{}'.format(Cov_XY))
print('sigma_X:\n{}'.format(sigma_X))
print('sigma_Y:\n{}'.format(sigma_Y))
print('corr(X,Y):\n{}'.format(r_XY))
```

```python
# example 2
X = np.array([20,15,13,10,7])
Y = np.array([60,55,50,45,40])
mu_x = np.mean(X)
mu_y = np.mean(Y)
X_zerooffset = X-mu_x
Y_zerooffset = Y-mu_y
Cov_XY = np.mean(X_zerooffset*Y_zerooffset)
sigma_X = np.mean(X_zerooffset**2)
sigma_Y = np.mean(Y_zerooffset**2)
r_XY = Cov_XY/np.sqrt(sigma_X*sigma_Y)
print('Cov_XY:\n{}'.format(Cov_XY))
print('sigma_X:\n{}'.format(sigma_X))
print('sigma_Y:\n{}'.format(sigma_Y))
print('corr(X,Y):\n{}'.format(r_XY))
```

## Generate data by multi-normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(202202)
def example_cov(mean, cov):
    data = np.random.multivariate_normal(mean, cov, 10000)
    plt.figure()
    plt.plot(data[:, 0], data[:, 1], 'x')
    plt.plot(mean[0],mean[1],'ko',markerfacecolor='white')
    plt.xlim([-8,8])
    plt.ylim([-8,8])
    plt.xlabel('x')
    plt.xlabel('y')
    plt.show()
    
# Example 1
mean = [0, 0]
cov = [[5, 0],
    [0, 2]]  # diagonal covariance 
example_cov(mean, cov)

# Example 2
mean = [0, 0]
cov = [[5, 1],
    [1, 2]]  # diagonal covariance 
example_cov(mean, cov)

# Example 3
mean = [0, 0]
cov = [[5, 3],
    [3, 2]]  # diagonal covariance
example_cov(mean, cov)
```
